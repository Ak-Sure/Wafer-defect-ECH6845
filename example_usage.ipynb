## Step 1: Import the Class

```python
from data_loading import WaferDataLoader
import matplotlib.pyplot as plt
import numpy as np
```

## Step 2: Initialize the DataLoader

This automatically loads the dataset and converts labels to class indices.

```python
# Initialize the loader
loader = WaferDataLoader()

print(f"Dataset loaded successfully!")
print(f"X shape: {loader.X.shape}")
print(f"y shape: {loader.y.shape}")
print(f"Number of classes: {loader.num_classes}")
```

## Step 3: Print Dataset Summary

```python
loader.print_summary()
```

## Step 4: View Class Distribution

```python
# Print class distribution table
loader.print_class_distribution(num_classes_to_show=10)
```

```python
# Plot class distribution
loader.plot_class_distribution()
```

## Step 5: View Pattern Table and Heatmap

```python
# Print all patterns
loader.print_pattern_table()
```

## Step 6: View Sample Gallery

```python
# Display one sample from each class
loader.plot_sample_gallery(samples_per_row=7)
```

## Step 7: Access Data Programmatically

```python
# Get a specific sample by class
class_idx = 5
wafer_map, pattern = loader.get_sample_by_class(class_idx, sample_num=0)

print(f"Class: {class_idx}")
print(f"Pattern: {pattern}")
print(f"Shape: {wafer_map.shape}")

plt.figure(figsize=(4, 4))
plt.imshow(wafer_map, cmap='viridis')
plt.title(f'Sample from Class {class_idx}\nPattern: {pattern}')
plt.colorbar()
plt.show()
```

## Step 8: Access Raw Data for Model Training

```python
# Now use the loaded data for your machine learning pipeline
X = loader.X
y = loader.y
classes_dict = loader.classes_dict
class_to_pattern = loader.class_to_pattern

print(f"Ready for model training!")
print(f"X: {X.shape}")
print(f"y: {y.shape}")
print(f"Classes: {loader.num_classes}")
```

## Summary of Available Methods

The `WaferDataLoader` class provides the following methods:

| Method | Purpose |
|--------|----------|
| `print_summary()` | Print dataset statistics and summary |
| `print_class_distribution()` | Print class distribution table |
| `plot_class_distribution()` | Visualize class distribution |
| `print_pattern_table()` | Print all 38 binary patterns |
| `plot_pattern_heatmap()` | Visualize patterns as heatmap |
| `plot_sample_gallery()` | Show one sample from each class |
| `get_sample_by_class()` | Get specific sample programmatically |

### Attributes

| Attribute | Description |
|-----------|-------------|
| `X` | Wafer maps array (38015, 52, 52) |
| `y` | Class labels (38015,) |
| `y_raw` | Raw binary patterns (38015, 8) |
| `classes_dict` | Pattern → Class mapping |
| `class_to_pattern` | Class → Pattern mapping |
| `num_classes` | Number of classes (38) |
| `path` | Dataset directory path |
